```python
X = df.drop('price', axis=1).values
y = df['price'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train.shape
```

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
model.add(Dense(128, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(
    x=X_train,
    y=y_train,
    validation_split=0.3,
    batch_size=128,
    epochs=400
)
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

```python
predictions = model.predict(X_test)
```

```python
with mlflow.start_run():
    print(f'Mean Absolute Error: {mean_absolute_error(y_test, predictions)}')
    print(f'Mean Squared Error: {mean_squared_error(y_test, predictions)}')
    print(f'Root Mean Squared Error: {mean_squared_error(y_test, predictions) ** 0.5}')
    print(f'Explained Variance Score: {explained_variance_score(y_test, predictions)}')
    
    rmse = (mean_squared_error(y_test, predictions) ** 0.5)
    
    mlflow.log_metrics({
        "mse": mean_squared_error(y_test, predictions),
        "rmse": rmse,
        "mae": mean_absolute_error(y_test, predictions),
        "evs": explained_variance_score(y_test, predictions)
    })
    
    mlflow.keras.log_model(model, 'model')
    
#     now = datetime.datetime.now()
#     modeldate = f'{now.year}-{now.month}-{now.day} {now.hour}:{now.minute}:{now.second}'
#     modelpath = f"../../artifacts/0"
#     mlflow.keras.save_model(model, modelpath)
```
